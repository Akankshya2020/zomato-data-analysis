```python
final.columns
```

```python
final.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size()
```

```python
ratings = final.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'rating count'})
ratings.head(5)
```

######observation


1. When rating is in between 4.5 to 4.9 ---> excellent
2. when rating is in between 4.0 to 4.4 ----> Very Good
3. when rating is in between 3.5 to 3.9 ---->Good
4. when rating is in between 3.0 to 3.4 ----> Average
5. when rating is in between 1.8 to 2.4 ----> poor and for 2148 orders there is no rating

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating",y="rating count",hue="Rating color",data=ratings,palette=['white','red','orange','yellow','green','green'])
```

observation:

 1. Not rated count is very high and the maximum number of rating are between 2.5 to 3.4

```python
import seaborn as sns
sns.countplot(x='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```
